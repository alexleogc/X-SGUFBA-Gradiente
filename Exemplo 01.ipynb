## <center> Determinação de minimos de uma equação polinomial utilizando o conceito de derivada</center>

```python
#Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

#### Suponha que desejamos obter a solução da Eq:

$$ f(x) = x^4 + 7x^3 + 5x^2 -17x + 3$$

```python
f = lambda x: x**4 + 7*x**3 + 5*x**2-17*x+3 # função
```

```python
plt.figure(figsize=(10,8))

x = np.linspace(-6.5,3)

plt.plot(x,f(x),label='$f(x)=x^4 + 7x^3 + 5x^2 -17x +3$') #Plotando a função f(x)

plt.ylabel('$f(x)$',fontsize=14)
plt.xlabel('$x$',fontsize=14)

plt.legend()
```

#### e sua derivada é dada por:

$$\frac{df}{dx} = 4x^3 + 21x^2 +10x -17$$

```python
df = lambda x: 4*x**3+21*x**2+10*x-17 #derivada da função a ser encontrada a solução
```

#### Se derivarmos a função $f$ em $x_0$, conseguiremos obter o vetor gradiente que aponta na direção de máximo da função. 

Suponha que queremos obter o gradiente em $x_0 = -5$:

```python
df(-5)
```

#### Para minimizar a função $f$, devemos tomar passos proporcionais a direção oposta do gradiente, ou seja:

$$x_{i+1} = x_i - \nabla f(x_i)$$

```python
df(-5)*(-1)
```

#### Repare que se utilizarmos o valor do gradiente em $x_i$, neste caso, a função irá caminhar diretamente para o ponto de máximo. Por esse motivo, devemos inserir um termo de proporção ($\alpha$):

$$x_{i+1} = x_i - \alpha \nabla f(x_i)$$

```python
n_int, x0, alpha = 10, -6, 0.001

xn = []

xn.append(x0)

for i in range(0,n_int):
    x0 = x0 - alpha*df(x0)
    xn.append(x0)
```

```python
plt.figure(figsize=(10,8))

plt.plot(x,f(x),label='$f(x)=x^4 + 7x^3 + 5x^2 -17x +3$') #Plotando a função f(x)

plt.ylabel('$f(x)$',fontsize=14)
plt.xlabel('$x$',fontsize=14)

aux=0
for i in xn:
    plt.scatter(i,f(i),label='iteração: %s' %aux)
    plt.text(i+0.1,f(i),str(aux))
    aux+=1
    
plt.legend()
```
